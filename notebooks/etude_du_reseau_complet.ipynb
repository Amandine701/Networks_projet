## 2.1 Constitution des rÃ©seaux


```python
# On regroupe les Ã©tablissements par catÃ©gorie (les plus frÃ©quentes)

col = df_check_in['location_type_name'].str.strip()

df_check_in['category_grouped'] = np.select(
    [

        #Restaurants (tous les noms contenant "Restaurant")
        col.str.contains("Restaurant|Ramen", case=False, na=False),

        # Retail (Shop / Store)
        col.str.contains("Shop|Store|Mall", case=False, na=False),

        # Food & Drink spÃ©cifiques
        col.isin(["CafÃ©", "Coffee Shop", "Bar"]),

        #  Transport
        col.isin(["Train Station", "Bus Station", "Subway"]),

        #  Services
        col.isin(["Office", "Medical Center"]),

        #  Outdoor / infrastructure
        col.isin(["Park", "Road", "Bridge"])

    ],
    [
        "FOOD_RESTAURANT",
        "RETAIL",
        "FOOD_DRINK",
        "TRANSPORT",
        "SERVICES",
        "OUTDOOR"
    ],
    default="OTHER"
)
```

```python
# CrÃ©ation des rÃ©seaux

def construction_reseau(dfjoint_complet, df_check_in):
    """
    dfagg_complet : contient la colonne 'checkins' (listes d'IDs) avec la liste de tous les check_ins par lieu
    df_check_in : contient 'location_id' et 'location_type_name'
    """
    df = dfjoint_complet.copy()

    # CrÃ©ation des arÃªtes
    df['edges'] = df['checkins'].apply(
        lambda x: list(itertools.combinations(x, 2))
    )
    edges = [edge for sublist in df['edges'] for edge in sublist]

    # CrÃ©ation du graphe
    G = nx.Graph()
    G.add_edges_from(edges)
    G.remove_edges_from(nx.selfloop_edges(G))

    # Ajout des attributs (type d'Ã©tablissement)
    dict_categories = dict(zip(df_check_in['location_id'], 
                               df_check_in['category_grouped']))
    
    # Assignation
    nx.set_node_attributes(G, dict_categories, 'category_grouped')

    return G
    

G_complet= construction_reseau(dfjoint_complet, df_check_in)
G_micro= construction_reseau(dfjoint_micro, df_check_in)
G_macro= construction_reseau(dfjoint_macro, df_check_in)
```

```python
def plot_stacked_degree_categories(G, n_bins=15, top_n_categories=23):
    # Extraire les donnÃ©es
    data = []
    for node, attrs in G.nodes(data=True):
        deg = G.degree(node)
        cat = attrs.get('category_grouped', 'Unknown') 
        data.append({'degree': deg, 'category_grouped': cat})
    
    df = pd.DataFrame(data)

    #  CrÃ©er les bins logarithmiques
    kmin, kmax = df['degree'].min(), df['degree'].max()
    if kmin <= 0: kmin = 1 
    
    # On crÃ©e les limites des bacs (bins)
    bins = np.logspace(np.log10(kmin), np.log10(kmax), num=n_bins)
    
    # Assigner chaque degrÃ© Ã  un bin (on utilise les intervalles comme labels)
    df['bin'] = pd.cut(df['degree'], bins=bins, include_lowest=True)
    
    #  Table croisÃ©e
    ct = pd.crosstab(df['bin'], df['category_grouped'], normalize='index')
    
    #  TracÃ©
    ax = ct.plot(kind='bar', stacked=True, figsize=(14, 7), width=0.8, colormap='Set3', edgecolor='black')
    
    # Graduation axe des abscisses : on garde l'intervalle complet mais on le formate (arrondi)
    new_labels = [f"[{int(b.left)}-{int(b.right)}]" for b in ct.index]
    
    ax.set_xticklabels(new_labels)
    
    plt.xticks(rotation=45, ha='right', fontsize=9)
    for label in ax.get_xticklabels():
        label.set_visible(True)
    # ---------------------------------
    plt.legend(
    title="Groupes",
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),
    ncol=4,   # ou 3 selon le nombre de catÃ©gories
    frameon=False
    )
    plt.tight_layout()
    plt.show()# Appel de la fonction
plot_stacked_degree_categories(G_complet)
plot_stacked_degree_categories(G_micro)
plot_stacked_degree_categories(G_macro)
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import itertools
from collections import Counter
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


# =========================================================
# 1. Construction du graphe pondÃ©rÃ© lieu-lieu
# =========================================================
def construction_reseau_weighted(dfjoint_complet, df_check_in):

    df = dfjoint_complet.copy()

    all_edges = []

    for x in df['checkins']:
        all_edges.extend(list(itertools.combinations(x, 2)))

    edge_weights = Counter(all_edges)

    G = nx.Graph()

    for (u, v), w in edge_weights.items():
        if u != v:
            G.add_edge(u, v, weight=w)

    # catÃ©gories lieux
    dict_categories = dict(zip(
        df_check_in['location_id'],
        df_check_in['category_grouped']
    ))

    nx.set_node_attributes(G, dict_categories, 'category_grouped')

    return G


# =========================================================
# 2. Ajout popularitÃ© moyenne des lieux
# =========================================================
def add_location_popularity(G, dfjoint_complet, df_users):

    user_pop = dict(zip(
        df_users['user_id'],
        df_users['nb_twitter_followers']
    ))

    df = dfjoint_complet.copy()

    def mean_pop(checkins):
        vals = [user_pop.get(u, np.nan) for u in checkins]
        vals = [v for v in vals if not np.isnan(v)]
        return np.mean(vals) if len(vals) > 0 else np.nan

    avg_pop = df.set_index('location_id')['checkins'].apply(mean_pop)

    nx.set_node_attributes(G, avg_pop.to_dict(), 'avg_popularity')

    return G


# =========================================================
# 3. Analyse rich-club adaptÃ© (degrÃ© vs popularitÃ©)
# =========================================================
def rich_club_analysis(G):

    deg = dict(G.degree())
    pop = nx.get_node_attributes(G, 'avg_popularity')

    common_nodes = [n for n in G.nodes if n in pop and not np.isnan(pop[n])]

    x = np.array([deg[n] for n in common_nodes])
    y = np.array([pop[n] for n in common_nodes])

    # corrÃ©lation globale
    corr, _ = pearsonr(x, y)

    print("CorrÃ©lation degrÃ© - popularitÃ© :", round(corr, 3))

    # scatter + fit
    plt.figure(figsize=(7,5))
    plt.scatter(x, y, alpha=0.4)

    # rÃ©gression simple
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x + b, color='red')

    plt.xlabel("DegrÃ© du lieu")
    plt.ylabel("PopularitÃ© moyenne (followers)")
    plt.title("Relation degrÃ© des lieux vs popularitÃ© des visiteurs")

    plt.tight_layout()
    plt.show()


# =========================================================
# 4. PIPELINE COMPLET
# =========================================================
G = construction_reseau_weighted(dfjoint_complet, df_check_in)
G = add_location_popularity(G, dfjoint_complet, df_users)
rich_club_analysis(G)
```
